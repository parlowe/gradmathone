### Taylor Swift and Travis Kelce: Modern Romance or the Best Marketing Ploy of the 21st Century?

**Introduction**

Taylor Swift has recently become one of the most popular figures in popular culture. Her Eras tour has broken records for both attendance, and revenue generated. However, fame has come at a price for the artisit, impacting the quality and stability of her romatic relationships. Most recently, her six-year relationship with British actor Joe Alwyn came to a close with speculation that the end was due to pressures from her career. 

However, Taylor has appeared to 'Shake Off' this relationship There are rumors ciruclating that Taylor has found a new match in Kansas City Chiefs tight end Travis Kelce. If the rumors are true Kelce would represent a divergence in the typical romantic partner Taylor has previously pursued. He is the first professional athlete in her reptoire, and has his own 'Reputation' in equal large franchise, the National Football League. 

On September 24th, 2023. Swift made an appearance at the Kansas City home game versus the Chicago Bears at Arrowhead stadium. This represented the clearest evidence to date of the possible pairing of Swift and Kelce, with Swift sitting in Kelce's personal box alongside his mother. As is expected, the public response was swift and concerted. The unexpected effect however was who this public included. For the first time, Swifties and NFL fans alike were buzzing with the latest Taylor Swift gossip. 



**Aims**

Is this bombshell romance actually what it is portrayed to be, or rather a complex marketing scheme by both parties to increase their popularity among target audiences that previously did not intersect? The aim of this project will be to invetigate this. Popularity for each of the parties will be quantified usng data supplied by Google for the relative interest score for Taylor Swift, Travis Kelce, and the Kansas City Chiefs before the first apperance of Swift on September 24th, and after. The effects of number of days since September 24th, number of Chiefs games that Swift has attended, and popularity in Kansas will be examined to qunatify the impact on each parties popularity. 

**Analysis**

The effects of days since September 24th, number of Chiefs games that Swift has attended, and relative popularity in Kansas will be examined to qunatify the impact on each party's popularity over time. Significant increases in popularity relative to the above factors will be indicators that popularity has been positvely influenced by the purported relationship. If there are no significant increases in popularity due to the factors studied or if significant predictor models cannot be generated, the conclusion will be that even if such a marketing scheme was proposed it did not have the intended effect of sustained increase in popularity. 

Output of this project will be dependant sample hypothesis testing using both a parametric and non-parametric test for each party will be completed using numerically constructed distribtuions. Regression models predicting the popularity of each party post 9/24 will be generated using the following effects: days since 9/24, chiefs games swift has attended since 9/24, and an effect for relative popularity of each party in Kansas since 9/24. The significance of each of these effects will be calcualted. Finally, a mixed effect parameter will also be considered in this model. 

**Data**

Data will be supplied by Google through their Google trends module. Here, relative interest is reported during a given timeframe based on search queries. Interest in normalized across the time frame studied with the highest interest being assigned a score of 100. The search terms "Taylor Swift", "Travis Kelce", and "Kansas City Chiefs" were used to generate relative interest CSVs segemented by week for the past 12 months. The terms were again used to generate relative interest data segmented on a daily basis for the days post September 24th. 

Finally, since the goal of this project is to elucidate how the intersection of two previously unrelated fan bases has affected the popularity of each party, the relative interest for each term in Kansas was generated. It is assumed that this represents a population most likely to be impacted by a potential relationship between Travis Kelce and Taylor Swift based on geographic location and home-team allegiance. This data was aslo segmented by day. 

It is important to note that the relative interest for datasets generated post September 24th will be updated daily until October 22nd to allow for more datapoints to be generated. 

```python
import pandas as pd 
from matplotlib import pyplot as pt
import numpy as np

population_data = np.array(pd.read_csv(r'C:\Users\eparl\Documents\Graduate Math\project_one_data.csv'))
post24data = np.array(pd.read_csv(r'C:\Users\eparl\Documents\Graduate Math\project_one_data_1.csv'))
post24data_kansas = np.array(pd.read_csv(r'C:\Users\eparl\Documents\Graduate Math\project_one_data_2.csv'))
post24data_games = np.array(pd.read_csv(r'C:\Users\eparl\Documents\Graduate Math\project_one_data_3.csv'))

x = np.arange(0,51,1)

y = np.arange(0,13,1)

pt.figure(0)
pt.plot(x, population_data[:,1], label = 'Taylor Swift')
pt.plot(x, population_data[:,2], label = 'Travis Kelce')
pt.plot(x, population_data[:,3], label = 'Kansas City Chiefs')
pt.xlabel('Week')
pt.ylabel('Relative Interest')
pt.title('Relative Interest Previous Year')
pt.legend()



pt.figure(1)
pt.plot(y, post24data[:,1], label = 'Taylor Swift')
pt.plot(y, post24data[:,2], label = 'Travis Kelce')
pt.plot(y, post24data[:,3], label = 'Kansas City Chiefs')
pt.xlabel('Day Since 9/24')
pt.ylabel('Relative Interest')
pt.title('Relative Interest Since 9/24')
pt.legend()

pt.figure(2)
pt.plot(y, post24data_kansas[:,1], label = 'Taylor Swift')
pt.plot(y, post24data_kansas[:,2], label = 'Travis Kelce')
pt.plot(y, post24data_kansas[:,3], label = 'Kansas City Chiefs')
pt.xlabel('Day Since 9/24')
pt.ylabel('Relative Interest')
pt.title('Relative Interest Since 9/24 - Kansas')
pt.legend()

pt.figure(3)
pt.plot(y, post24data_games[:,1])
pt.xlabel('Day Since 9/24')
pt.ylabel('Number of Games Attended')
pt.title('Number of Games Attended since 9/24')
```
